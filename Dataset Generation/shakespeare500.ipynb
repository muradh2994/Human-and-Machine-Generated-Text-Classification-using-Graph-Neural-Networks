```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# load the shakespeare data
data = pd.read_csv('Shakespeare_data.csv',index_col=False)
data
```

```python
# filter the content column to 500 words in each line 
df_shakespeare = data['PlayerLine'].groupby(np.repeat(np.arange(len(data)), 50)[:len(data)]).agg(' '.join)
df_shakespeare
```

```python
df = pd.DataFrame(df_shakespeare)
```

```python
# consider only 500 values
df = df.head(500)
df
```

```python
# Assing label as human
df = df.assign(label='Human')
```

```python
# Column Renames
df = df.rename(columns={'PlayerLine': 'text',
                   'label': 'Label'})
df
```

```python
# Save as csv file
df.to_csv('Human_shakespeare.csv',index=False)
```
